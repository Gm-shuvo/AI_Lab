```python
def sigmoid_numpy(X):
   return 1/(1+np.exp(-X))

sigmoid_numpy(np.array([12,0,1]))
```

```python
def log_loss(y_true, y_prediction):
  epsilon = 1e-15
  y_prediction_new = [max(i, epsilon) for i in y_prediction]
  y_prediction_new = [min(i, 1 - epsilon) for i in y_prediction]
  y_prediction_new = np.array(y_prediction_new)
  return -np.mean(y_true* np.log(y_prediction_new) + (1 - y_true)* np.log(1 - y_prediction_new))

```

```python
def gradient_descent(age, affordibility, y_true, epchos, loss_thresold):
  w1 = w2 = 1
  bias = 0
  learning_rate = 0.5
  
  # print(y_prediction)
  age = age.astype('float32')
  affordibility = affordibility.astype('float32')
  n = len(age)
  for i in range(epchos):
    weight_sum = w1 * age + w2 * affordibility + bias
    y_prediction = sigmoid_numpy(weight_sum)
    loss = log_loss(y_true, y_prediction)
    # print(y_prediction)
    w1d = (1/ n) * np.dot(np.transpose(age), ( y_prediction - y_true))
    w2d = (1 / n) * np.dot(np.transpose(affordibility), (y_prediction - y_true))
    bias_d = np.mean(y_prediction - y_true)

    w1 = w1 - learning_rate * w1d
    w2 = w2 - learning_rate * w2d
    bias = bias - learning_rate * bias_d

    print(f'Epoch: {i}, w1: {w1}, w2: {w2}, bias:{bias}, loss:{loss}')
    if loss <= loss_thresold:
      break
  return w1, w2, bias


```

```python
#define MyCustomModel
class MyNN():
  def __int__(self):
    self.w1 = 1
    self.w2 = 1
    self.bias = 0
  

  def fit(self, X, y, epchos, loss_thresold):
    self.w1,self.w2, self.bias = self.gradient_descent(X['age'],X['affordibility'], y, epchos, loss_thresold)

  def predict(self, x_test):
    weight_sum = self.w1 * x_test['age'] + self.w2 * x_test['affordibility'] + self.bias
    return sigmoid_numpy(weight_sum)


  def gradient_descent(self, age, affordibility, y_true, epchos, loss_thresold):
    w1 = w2 = 1
    bias = 0
    learning_rate = 0.5
    
    # print(y_prediction)
    age = age.astype('float32')
    affordibility = affordibility.astype('float32')
    n = len(age)
    for i in range(epchos):
      weight_sum = w1 * age + w2 * affordibility + bias
      y_prediction = sigmoid_numpy(weight_sum)
      loss = log_loss(y_true, y_prediction)
      # print(y_prediction)
      w1d = (1/ n) * np.dot(np.transpose(age), ( y_prediction - y_true))
      w2d = (1 / n) * np.dot(np.transpose(affordibility), (y_prediction - y_true))
      bias_d = np.mean(y_prediction - y_true)

      w1 = w1 - learning_rate * w1d
      w2 = w2 - learning_rate * w2d
      bias = bias - learning_rate * bias_d
      if i%50 == 0:
        print(f'Epoch: {i}, w1: {w1}, w2: {w2}, bias:{bias}, loss:{loss}')
      if loss <= loss_thresold:
        break
    return w1, w2, bias
```
